### *Describe* method to obtain statistical information about the dataset

```python
data = red_wine.describe()
data
```

From the info method section, null values are present in some columns and hence calls for ***data preprocessing***.
Therefore these null values can be replaced by the mean values of their respective columns which we obtianed using the ***describe*** method.
The mean values can be obtained using the ***loc*** method and then type casting them into a list.

### Preprocessing

```python
mean = list(data.loc['mean'])
mean
```

The ***fillna*** method is used to fill the null values with their mean, the column is reassigned since the ***fillna*** method returns the whole 
column after filling it with the mean value.

```python
red_wine['fixed acidity'] = red_wine['fixed acidity'].fillna(mean[0])
```

```python
red_wine['volatile acidity'] = red_wine['volatile acidity'].fillna(mean[1])
```

```python
red_wine['chlorides'] = red_wine['chlorides'].fillna(mean[4])
```

```python
red_wine['free sulfur dioxide'] = red_wine['free sulfur dioxide'].fillna(mean[5])
```

```python
red_wine['density'] = red_wine['density'].fillna(mean[7])
```

```python
red_wine['quality'] = red_wine['quality'].fillna(mean[11])
```

```python
red_wine.info()
```

Using the ***info*** method, it can be seen that all the columns have a non-Null value count of 1599 which is corresponding to the number of entries or indexes.

Duplicated data can be checked using the ***duplicated*** method

### Ploting graphs

```python
from matplotlib import pyplot as plt 
```

A bar chart can be plotted between the alcohol column and the quality column to visualize their relationship.

```python
plt.figure(figsize=[10,6])
plt.bar(red_wine['quality'],red_wine['alcohol'])
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

The bar chart above indicates that most wines have higher quality if their alcohol value is within the ranges of 12 and 14.

```python
corr = red_wine.corr() 
```

A correlation model can be achieved using the ***corr*** method.
The correlation model can be visualized using a module known as ***statsmodel***, which is built on top of matplotlib.

```python
import statsmodels.api as sm
```

```python
sm.graphics.plot_corr(corr,xnames=list(corr.columns))
```

In the above graph, shades of blue indicate lower correlation and shades of brown indicate a much higher correlation.
It is seen that the alcohol column has a much higher correlation with the quality column than any other column.
The graph can be further analyzed to be able to predict the quality of wines made.
